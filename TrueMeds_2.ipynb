```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualize count of missing values per column
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=missing_values.index, y=missing_values.values)
plt.xticks(rotation=90)
plt.title('Count of Missing Values per Column')
plt.show()
```

```python
df['Event Revenue'] = df['Event Revenue'].fillna(df['Event Revenue'].median())

# Filling with 'None' or another placeholder for categorical columns
df['Operator'] = df['Operator'].fillna('None')
df['Carrier'] = df['Carrier'].fillna('None')
df['Language'] = df['Language'].fillna('None')
df['Device Category'] = df['Device Category'].fillna('None')
df['OS Version'] = df['OS Version'].fillna('None')
df['SDK Version'] = df['SDK Version'].fillna('None')
df['App Name'] = df['App Name'].fillna('None')
df['Device Model'] = df['Device Model'].fillna('None')
df['State'] = df['State'].fillna('None')
df['City'] = df['City'].fillna('None')
```

```python
df.isnull().sum()
```

```python
# Convert time columns to datetime
df['Install Time'] = pd.to_datetime(df['Install Time'])
df['Event Time'] = pd.to_datetime(df['Event Time'])
```

```python
df.head()
```

```python
# Calculate time differences
df['Time_to_Event'] = (df['Event Time'] - df['Install Time']).dt.total_seconds() / 3600  # in hours
```

```python
# Create separate dataframes for each event
df_app_order_placed = df[df['Event Name'] == 'app_order_placed']
df_box_verified = df[df['Event Name'] == 'box_verified']
df_order_delivered = df[df['Event Name'] == 'order_delivered']
```

```python
# Aggregate by Media Source
avg_time_to_event = df.groupby('Media Source')['Time_to_Event'].mean().reset_index()
```

```python
avg_time_to_event
```

```python
# Plot average time to event by Media Source
plt.figure(figsize=(12, 8))
sns.barplot(data=avg_time_to_event, x='Media Source', y='Time_to_Event', palette='Dark2')
plt.title('Average Time to Event by Media Source')
plt.xlabel('Media Source')
plt.ylabel('Average Time to Event (hours)')
plt.show()
```

```python
# Distribution of time delta between install and event times
plt.figure(figsize=(10, 6))
sns.histplot(df['Time_to_Event'], kde=True)
plt.title('Distribution of Time Delta Between Install and Event Times')
plt.xlabel('Time Delta (hours)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Revenue by Media Source
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Media Source', y='Event Revenue', estimator=sum, ci=None)
plt.title('Revenue by Media Source')
plt.xlabel('Media Source')
plt.ylabel('Total Revenue')
plt.show()
```

```python
# Event Revenue by Location
coupon_usage_location = df.groupby('State')['Event Revenue'].mean().reset_index()
plt.figure(figsize=(14, 8))
sns.barplot(data=coupon_usage_location, x='State', y='Event Revenue')
plt.title('Event Revenue by Location')
plt.xlabel('State')
plt.ylabel('Average Event Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate time differences for each funnel stage
df['Time_to_Box_Verified'] = df.groupby('Campaign ID')['Time_to_Event'].shift(-1) - df['Time_to_Event']
df['Time_to_Order_Delivered'] = df.groupby('Campaign ID')['Time_to_Event'].shift(-2) - df['Time_to_Event']
```

```python
# Calculate the delta for each event and media source
df_event_time = df.groupby(['Media Source', 'Event Name'])['Time_to_Event'].mean().reset_index()
```

```python
# Find the media source with the biggest delta
media_source_delta = df_event_time.groupby('Media Source')['Time_to_Event'].mean().reset_index()
biggest_delta_media_source = media_source_delta.loc[media_source_delta['Time_to_Event'].idxmax()]
```

```python
# Calculate average time from install to each event
average_time_to_events = df.groupby('Event Name')['Time_to_Event'].mean().reset_index()
```

```python
# Visualization
plt.figure(figsize=(12, 6))
sns.barplot(data=media_source_delta, x='Media Source', y='Time_to_Event')
plt.title('Average Time to Event by Media Source')
plt.xlabel('Media Source')
plt.ylabel('Average Time to Event (hours)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=average_time_to_events, x='Event Name', y='Time_to_Event')
plt.title('Average Time to Event')
plt.xlabel('Event Name')
plt.ylabel('Average Time to Event (hours)')
plt.show()
```

```python
print("Media Source with the biggest delta: ", biggest_delta_media_source)
print("Average time to events:\n", average_time_to_events)
```

```python
# 4) Top Campaigns to Increase Spending
```

```python
# Calculate total revenue and conversion rates by Campaign ID
campaign_performance = df.groupby('Campaign ID').agg({
    'Event Revenue': 'sum',
    'is_core_customer': 'sum',
    'Event Name': 'count'  # Assuming count of events as a proxy for engagement
}).reset_index()
```

```python
# Calculate conversion rate
campaign_performance['Conversion Rate'] = campaign_performance['is_core_customer'] / campaign_performance['Event Name']

# Sort campaigns by revenue and conversion rate
top_campaigns = campaign_performance.sort_values(by=['Event Revenue', 'Conversion Rate'], ascending=False).head(10)
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(data=top_campaigns, x='Campaign ID', y='Event Revenue', palette='viridis')
plt.title('Top Campaigns by Revenue')
plt.xlabel('Campaign ID')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
print("Top campaigns to increase spending:\n", top_campaigns)
```
